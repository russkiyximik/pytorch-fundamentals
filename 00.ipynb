### Matrix manipulation

```python
tensor = torch.tensor([1,2,3])
tensor + 10
```

```python
tensor * 10, tensor - 10, torch.mul(tensor, 10), torch.add(tensor, 10)
```

```python
# Manual matrix multiplication
a = torch.rand(3,3)
b = torch.rand(3,3)
c = torch.zeros(3,3)

# c = a * b

for i in range(a.shape[0]):
    for j in range(a.shape[1]):
        c[i][j] = a[i][j] * b[j][i] # element-wise

c = torch.zeros(3,3)

for i in range(a.shape[0]):
    for j in range(b.shape[1]):
        for k in range(c.shape[0]):
            c[i][j] += a[i][k] * b[k][j] # matrix multiplication
a, b, c
```

```python
# Equivalently:
c = torch.matmul(a, b)
c
```

Rules for matrix multiplication:
1. The **inner dimensions** must match, i.e. `(m,n) * (n,m)`
2. The resulting matrix has the shape of the **outer dimensions**,
i.e. `(m,n) * (n,m)` -> `(n,n)`

```python
torch.matmul(torch.tensor([[1,2],[1,2],[1,2]]), torch.tensor([[2,3,4],[2,3,4]]))
```

```python
torch.tensor([[1,2],[1,2],[1,2]]).shape, torch.tensor([[2,3,4],[2,3,4]]).shape
```

```python
# Shapes for matrix multiplication
A = torch.tensor([[1, 2],
                  [3, 4],
                  [5, 6]])

B = torch.tensor([[7, 10],
                  [8, 11],
                  [9, 12]]).T

# Manual transpose
C = torch.zeros_like(B.T) # lil shortcut to flip dimensions
for i in range(B.shape[0]):
    for j in range(B.shape[1]):
        C[j][i] = B[i][j]

torch.mm(A, B) # .mm() is the same as .matmul()
```
