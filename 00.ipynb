### Matrix manipulation

````python
tensor = torch.tensor([1,2,3])
tensor + 10
````

````python
tensor * 10, tensor - 10, torch.mul(tensor, 10), torch.add(tensor, 10)
````

````python
# Manual matrix multiplication
a = torch.rand(3,3)
b = torch.rand(3,3)
c = torch.zeros(3,3)

# c = a * b

for i in range(a.shape[0]):
    for j in range(a.shape[1]):
        c[i][j] = a[i][j] * b[j][i] # element-wise

c = torch.zeros(3,3)

for i in range(a.shape[0]):
    for j in range(b.shape[1]):
        for k in range(c.shape[0]):
            c[i][j] += a[i][k] * b[k][j] # matrix multiplication
a, b, c
````

````python
# Equivalently:
c = torch.matmul(a, b)
c
````

Rules for matrix multiplication:
1. The **inner dimensions** must match, i.e. `(m,n) * (n,m)`
2. The resulting matrix has the shape of the **outer dimensions**,
i.e. `(m,n) * (n,m)` -> `(n,n)`

````python
torch.matmul(torch.tensor([[1,2],[1,2],[1,2]]), torch.tensor([[2,3,4],[2,3,4]]))
````

````python
torch.tensor([[1,2],[1,2],[1,2]]).shape, torch.tensor([[2,3,4],[2,3,4]]).shape
````

````python
# Shapes for matrix multiplication
A = torch.tensor([[1, 2],
                  [3, 4],
                  [5, 6]])

B = torch.tensor([[7, 10],
                  [8, 11],
                  [9, 12]]).T

# Manual transpose
C = torch.zeros_like(B.T) # lil shortcut to flip dimensions
for i in range(B.shape[0]):
    for j in range(B.shape[1]):
        C[j][i] = B[i][j]

torch.mm(A, B) # .mm() is the same as .matmul()
````

Reshaping - reshpaes an input tensor to a defined shape

View - return a view of an input tensor of a certain shape but keeps the same memory as the original tensor

Stacking - combines multiple tensor on top of (vstack) or next to (hstack) each other

Squeeze - removes all `1` dimensions from a tensor

Unsqueeze - add a `1` dimension to a tensor

Permute - return a view of the input with dimensions permuted in a certain way

````python
x = torch.arange(1., 10.)
x, x.shape
````

````python
x_reshaped = x.reshape(3,3)
x_reshaped, x_reshaped.shape
````

````python
z = x.view(1,9)
# Changing z changes x (z and x share the same loc in memory)
z[:, 0] = 5
z, x
````

````python
x_stacked = torch.stack([x,x,x,x], dim=0)
x_stacked, torch.stack([x,x,x,x], dim=1)
````

````python
torch.vstack([x,x,x,x])
````

````python
x_reshaped = x_reshaped.reshape(1,9)
x_reshaped, x_reshaped.squeeze()
````

````python
x_squeezed = x_reshaped.squeeze()
x_unsqueezed = x_squeezed.unsqueeze(dim=0)
x_unsqueezed
````

## PyTorch tensors & NumPy

Data in NumPy, want in PyTorch tensor -> `torch.from_numpy(ndarray)`

PyTorch tensor -> NumPy -> `torch.Tensor.numpy()`

````python
import numpy as np

array = np.arange(1.0, 8.0) # float64 default
tensor = torch.from_numpy(array) # .type(torch.float32)
array, tensor
````

````python
array = array + 1
array, tensor
````

````python
tensor = torch.ones(7)
numpy_tensor = tensor.numpy()
tensor, numpy_tensor
````

````python
tensor = tensor + 1
tensor, numpy_tensor
````
