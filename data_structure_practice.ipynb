## Linked List (single)

```python
class Node:
    def __init__(self, val):
        self.val = val
        self.next: Node | None = None
```

```python
a = Node(5)
b = Node(4)
c = Node(3)
```

```python
class LinkedList:
    def __init__(self):
        self.head: Node | None = None

    def insert_at_beginning(self, node: Node):
        node.next = self.head
        self.head = node

    def append(self, node: Node):
        if not self.head: # Checks if list is empty
            self.insert_at_beginning(node)
            return
        current = self.head # could implement using ptr
        while current.next:
            current = current.next # loops through list
        current.next = node

    def insert(self, node: Node, idx: int):
        if self.head is None or idx <= 0:
            self.insert_at_beginning(node)
            return
        # idx between 0 (beginning of list) and list length (end of list)
        # Must search list to find length - O(n)
        length = 1
        current = self.head
        while current.next:
            current = current.next
            length += 1
        if idx > length:
            idx = length

        current = self.head
        for i in range(idx - 1):
            if current.next is None: break
            current = current.next
        node.next = current.next
        current.next = node
        

    def view(self):
        if self.head is None: return
        current = self.head
        while current.next:
            print(f"{current.val} -> ", end='')
            current = current.next
        print(current.val)
            
```

```python
l = LinkedList()
l.append(b)
l.append(a)
l.insert_at_beginning(c)
l.insert(Node(5), 1)
l.view() # Should print 3 -> 5 -> 4 -> 5
```

## Stack

```python
class Stack:
    def __init__(self):
        self.top: Node | None = None
    
    def append(self, node: Node):
        node.next = self.top
        self.top = node
    
    def pop(self):
        if self.top is None: return
        self.top = self.top.next

    def view(self):
        if self.top is None: return
        current = self.top
        while current.next:
            print(f"{current.val} -> ", end='')
            current = current.next
        print(current.val)
            
```

```python
a = Node(3)
b = Node(5)
c = Node(7)
s = Stack()
s.append(c)
s.append(b)
s.pop()
s.append(a)
s.view()
```

## Queue

```python
class Queue:
    def __init__(self):
        self.start: Node | None = None
        self.end: Node | None = None
    
    # Nodes should point start to end. Otherwise, we would have to 
    # find the node before self.start to set its .next equal to None

    def push(self, node: Node):
        if self.end is None:
            self.start = self.end = node
            return
        self.end.next = node
        self.end = node
    
    def pop(self):
        if self.start == self.end or self.start is None: self.start = self.end = None; return
        # Pop from the start
        self.start = self.start.next

    def view(self):
        if self.start is None: return
        current = self.start
        while current.next:
            print(f"{current.val} -> ", end='')
            current = current.next
        print(current.val)
    
```

```python
q = Queue()
q.push(Node(1))
q.push(Node(2))
q.push(Node(1))
q.pop()
q.view()
```
